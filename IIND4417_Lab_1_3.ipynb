#Limpieza de datos

La limpieza de datos significa corregir datos. Los datos incorrectos podrían ser:

* Celdas vacías.
* Datos en formato incorrecto.
* Información incorrecta.
* Duplicados.


````python
import pandas as pd
````

````python
df = pd.read_csv('/content/drive/MyDrive/Datasets/calories.csv')
df.head()
````

````python
df.tail()
````

````python
print("Número total de registros: ", len(df.index),"\n")
print(df.to_string())
````

## Eliminar filas
Eliminar todas las filas con valores NULL.

````python
# copia el data frame
df2 = df.copy()
df2.dropna(inplace = True)

print("Número total de registros: ", len(df2.index),"\n")
print(df2.to_string())
````

##Reemplazar valores vacíos

Otra forma de tratar con celdas vacías es insertar un nuevo valor.

* De esta manera, no tendrá que eliminar filas enteras solo por tener algunas celdas vacías.

* El método ```fillna()``` nos permite reemplazar celdas vacías con un valor.

````python
# copia el data frame
df3 = df.copy()
df3.fillna(12500, inplace = True)

print("Número total de registros: ", len(df3.index),"\n")
print(df3.to_string())
````

##Reemplazar solo columnas especificadas

El ejemplo anterior reemplaza todas las celdas vacías en todo el marco de datos.

* Para reemplazar solo valores vacíos en una columna.
* Se debe especificar el nombre de la columna.

````python
# copia el data frame
df4 = df.copy()
df4["Calories"].fillna(18000, inplace = True)

print("Número total de registros: ", len(df4.index),"\n")
print(df4.to_string())

# no se remplazo el NaN en Date del registro 22
````

##Reemplazar usando media, mediana o moda

* Una forma común de reemplazar celdas vacías es calcular el valor medio, mediano o moda de la columna.

* Pandas utiliza los métodos ```mean()```, ```mediana()```, y ```mode()``` para calcular los valores respectivos para una columna específica.

````python
# copia el data frame
df5 = df.copy()

mediaCalorias = df5["Calories"].mean()
df5["Calories"].fillna(mediaCalorias, inplace = True)

print("Número total de registros: ", len(df5.index),"\n")
print(df5.to_string())

# no se puede extraer la media de Date por ser cadena de caracteres
````

````python
# copia el data frame
df6 = df.copy()

medianaCalorias = df6["Calories"].median()
df6["Calories"].fillna(medianaCalorias, inplace = True)

print("Número total de registros: ", len(df6.index),"\n")
print(df6.to_string())

# noten como el valor de la mediana es diferente que el de la media
````

````python
# copia el data frame
df7 = df.copy()

medianaCalorias = df7["Calories"].mode()[0]
df7["Calories"].fillna(medianaCalorias, inplace = True)

print("Número total de registros: ", len(df7.index),"\n")
print(df7.to_string())
````
